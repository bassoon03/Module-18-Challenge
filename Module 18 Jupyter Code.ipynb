```python
# Dependencies: pandas and datetime
import pandas as pd
from datetime import datetime
```

```python
#  Read in CitiBike file path for April 2023
new_file = pd.read_csv('Module 18 Challenge\JC-202304-citibike-tripdata.csv')
```

```python
# Put csv file into DataFrame
new_file_df = pd.DataFrame(new_file)
new_file_df
```

```python
# Check data types of data frame
new_file_df.dtypes
```

```python
# Convert started_at and ended_at columns to datatime data type
new_file_df['started_at'] = pd.to_datetime(new_file_df['started_at'])
new_file_df['ended_at'] = pd.to_datetime(new_file_df['ended_at'])
```

```python
# Check that started_at and ended_at columns were converted to datetime
new_file_df.dtypes
```

```python
# Create a new column for Trip Duration by calculating difference between ended_at and started_at.
# Value given in seconds.
# Form of this code sourced from Saturn Cloud
new_file_df['Trip Duration (seconds)'] = (new_file_df["ended_at"] - new_file_df["started_at"]).dt.seconds
new_file_df
```

```python
# Sort the data frame by values in started_at column
new_file_df.sort_values(by = ['started_at'], inplace = True)
new_file_df
```

```python
# Eliminate all rows where Trip Duration is 60 seconds or less.
# Supposedly these had already been taken out to eliminate potential false starts and redocking to ensure that a bike is secure.
# This does not appear to be the case, at least not with this data set.
new_file_df = new_file_df[new_file_df['Trip Duration (seconds)'] > 60]
new_file_df
```

```python
# Drop rows with missing values and reset index
new_file_df.dropna(how = 'any', inplace = True)
new_file_df = new_file_df.reset_index()
```

```python
# Popularity of each station as a start and end point calculated
start_popularity = new_file_df["start_station_name"].value_counts()
end_popularity = new_file_df["end_station_name"].value_counts()

# start_popularity and end_popularity converted to data frames
start_popularity = pd.DataFrame(start_popularity)
end_popularity = pd.DataFrame(end_popularity)

# start_popularity and end_popularity data frames merged to new_file_df


new_file_df = new_file_df.merge(start_popularity, how = 'left', on = "start_station_name")
new_file_df = new_file_df.merge(end_popularity, how = 'left', on = "end_station_name")


# Columns that are added on are titled count_x and count_y.
# Rename the columns as Start Station Popularity and End Station Popularity respectively
new_file_df = new_file_df.rename(columns = {"count_x":"Start Station Popularity", "count_y":"End Station Popularity"})

new_file_df
```
