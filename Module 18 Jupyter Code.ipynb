```python
import pandas as pd
from datetime import datetime
```

```python
file1 = pd.read_csv('Module 18 Challenge\JC-202401-citibike-tripdata.csv')
file2 = pd.read_csv('Module 18 Challenge\JC-202402-citibike-tripdata.csv')
file3 = pd.read_csv('Module 18 Challenge\JC-202403-citibike-tripdata.csv')
```

```python
jan_2024_df = pd.DataFrame(file1)
feb_2024_df = pd.DataFrame(file2)
mar_2024_df = pd.DataFrame(file3)
```

```python
jan_2024_df.head()
```

```python
feb_2024_df.head()
```

```python
mar_2024_df.head()
```

```python
quarter1_df = pd.concat([jan_2024_df, feb_2024_df, mar_2024_df])
quarter1_df
```

```python
quarter1_df.dtypes
```

```python
quarter1_df['started_at'] = pd.to_datetime(quarter1_df['started_at'])
quarter1_df['ended_at'] = pd.to_datetime(quarter1_df['ended_at'])
```

```python
quarter1_df.dtypes
```

```python
# Saturn Cloud
quarter1_df['Trip Duration (seconds)'] = (quarter1_df["ended_at"] - quarter1_df["started_at"]).dt.seconds
quarter1_df
```

```python
quarter1_df.sort_values(by = ['started_at'], inplace = True)
quarter1_df
```

```python
quarter1_df = quarter1_df[quarter1_df['Trip Duration (seconds)'] > 60]
quarter1_df
```

```python
quarter1_df.dropna(how = 'any', inplace = True)
quarter1_df = quarter1_df.reset_index()
```

```python
start_popularity = quarter1_df["start_station_name"].value_counts()
end_popularity = quarter1_df["end_station_name"].value_counts()

start_popularity = pd.DataFrame(start_popularity)
end_popularity = pd.DataFrame(end_popularity)

# quarter1_df = quarter1_df.merge(start_popularity, how = 'left', on = "start_station_name")
# quarter1_df = quarter1_df.merge(end_popularity, how = 'left', on = "end_station_name")


quarter1_df = quarter1_df.rename(columns = {"count_x":"Start Station Popularity", "count_y":"End Station Popularity"})

quarter1_df

```
